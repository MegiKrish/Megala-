```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
car=pd.read_csv('cars.csv')
```

```python
car
```

## Bar Plot Graph

```python
plt.bar(car.Carb,car.Gear,color='red')
plt.xlabel('Number of carburetors')
plt.ylabel('Number of forward Gears')
plt.title('Carbs vs Gear')
plt.show()
```

```python
plt.barh(car.Carb,car.Gear,color='blue')
plt.xlabel('Number of carburetors')
plt.ylabel('Number of forward Gears')
plt.title('Carbs vs Gear')
plt.show()
```

```python
car.plot(x = 'Carb', y = 'Gear' , kind = 'bar')
plt.show()
```
