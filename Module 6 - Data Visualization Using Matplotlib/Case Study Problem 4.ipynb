## 1.Bar Plot for Cylinder

```python
import matplotlib.pyplot as plt
import pandas as pd
car=pd.read_csv('cars.csv')
plt.bar(car.Cylinder,car.Qsec,color='red')
plt.xlabel('Number of Cylinders')
plt.ylabel('Frequency of Cars')
plt.title('Number of Cylinders vs Frequency of Cars')
plt.show()
```

## 2.Bar Graph with null values

```python
data=car.isnull().sum()
```

```python
car=pd.read_csv('cars.csv')
plt.figure(figsize=(12,7))
plt.bar(car.columns,data)
plt.xlabel('Cars')
plt.ylabel('Count')
plt.title('Bar Graph with Null values')
plt.show()
```

## 3.Histogram for MPG

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.hist(car.MPG,bins=30,color='olive')
plt.xlabel('Miles per Gallon')
plt.ylabel('Frequency')
plt.title('Miles Per Gallon Histogram')
plt.show()
```

## 4.Box Plot for HP

```python
plt.figure(figsize=(15,7))
plt.boxplot(car.HP)
plt.xlabel(' Car Horse Power')
plt.title('Box Plot for Car Horse Power')
plt.show()
```
